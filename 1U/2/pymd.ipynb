#Cargar Librerias

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

#Consumir Datos

```python
data = pd.read_csv("https://raw.githubusercontent.com/rpizarrog/Analisis-Inteligente-de-datos/main/datos/datos.FIFA.limpios.csv", encoding= 'unicode_escape')
```

#Explorar Datos

```python
data.describe()
```

Top 5 juadores peor pagados

```python
data.sort_values(by=['Valor'], ascending=True).head(5)
```

Nacion con jugadores con menor valor

```python
data.groupby('Nationality')['Valor'].mean().sort_values(ascending=True).head(1)
```

Conteo de Jugadores que prefieren el pie para patear

#Visualizar Datos

```python
data.groupby(['Preferred.Foot'])['Preferred.Foot'].count().plot(kind='pie')
```
