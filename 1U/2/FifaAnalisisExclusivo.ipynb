#Objetivo
Analisar, explorar datos, para recaudar variables de interes, los cuales nos permiten tener una mejor compresion de los jugadores segun sus estadisticas

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

Cargar datos

```python
data = pd.read_csv("https://raw.githubusercontent.com/rpizarrog/Analisis-Inteligente-de-datos/main/datos/datos.FIFA.limpios.csv", encoding= 'unicode_escape')
```

Describir los datos a explorar los primeros 10 registros

```python
data.head(10)
```

Describir estructura de datos

```python
data.describe()
```

## ¿Qué jugadores ganan más dinero, el top 5?

```python
data.sort_values(by=['Valor'], ascending=False).head(5)
```

¿Cuántos jugadores de fútbol son izquierdos?

```python
data.groupby(['Preferred.Foot'])['Preferred.Foot'].count()['Left']
```

¿Cuál es el valor medio (promedio) de estaura en cms o metros de jugadores de fútbol?

```python
data[['Estatura']].mean()
```

¿Cómo se llamna los jugadores más bajitos en estatura?


```python
data[['Name','Estatura']].sort_values(by=['Estatura']).head(10)
```

De qué nacionalidad o país hay mas jugadres registrados?


```python
data.groupby('Nationality')['Nationality'].count().sort_values(ascending=False).head(1)
```
